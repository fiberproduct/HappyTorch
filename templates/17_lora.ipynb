Implement **LoRA** for parameter-efficient fine-tuning of large models.

LoRA adds a low-rank decomposition to existing weights:
$$W' = W + \Delta W = W + \frac{\alpha}{r}BA$$

Where $B \in \mathbb{R}^{d_{out} \times r}$ and $A \in \mathbb{R}^{r \times d_{in}}$.


### Requirements
- Initialize B with zeros (so LoRA starts as identity)
- Initialize A with Kaiming uniform or normal
- Forward: `x @ W.T + (x @ A.T @ B.T) * alpha / rank`
