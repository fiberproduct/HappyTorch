# 🔥 TorchCode — PyTorch Operator Practice

Welcome! This is a JupyterLab-based environment for practicing PyTorch operator implementations.

## 📁 File Layout

Each problem has **two** notebooks:

| File | Description |
|------|-------------|
| `01_relu.ipynb` | ✏️ **Blank template** — write your implementation here (resets on every startup) |
| `01_relu_solution.ipynb` | 📖 **Reference solution** — check when you're stuck |

> ⚠️ Blank templates are reset to their initial state on every `make run`.  
> To keep your code, save it under a different filename.

## How to Use

1. **Open a problem** — pick a blank template like `01_relu.ipynb`
2. **Read the description** — the first cell has the problem statement and signature
3. **Write your code** — implement in the `✏️ YOUR IMPLEMENTATION HERE` cell
4. **Debug** — insert cells freely for `print(x.shape)` etc.
5. **Submit** — run the last cell `check("relu")` to see results
6. **View solution** — open the corresponding `_solution.ipynb`
