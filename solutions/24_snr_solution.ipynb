# 🟡 Solution: Sigmoid Noise Schedule

```python
import torch
```

```python
# ✅ SOLUTION

def sigmoid_schedule(t: torch.Tensor, beta_start: float, beta_end: float, gamma: float = 10.0) -> torch.Tensor:
    # t: normalized timesteps in [0, 1]
    # Sigmoid schedule creates smooth S-curve transition
    
    # Center the sigmoid around 0.5
    sig = torch.sigmoid(gamma * (t - 0.5))
    
    # Scale to [beta_start, beta_end]
    betas = beta_start + (beta_end - beta_start) * sig
    
    return betas
```

```python
# Verify
t = torch.linspace(0, 1, 100)
betas = sigmoid_schedule(t, beta_start=0.0001, beta_end=0.02)
print(f"Beta at t=0: {betas[0]:.6f}")
print(f"Beta at t=1: {betas[-1]:.6f}")
print(f"Beta at t=0.5: {betas[49]:.6f}")
```
