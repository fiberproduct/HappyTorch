# 🟡 Solution: GELU Activation

Reference solution for GELU using the exact formula with erf.

```python
import torch
import math
```

```python
# ✅ SOLUTION (Exact formula)

def gelu(x: torch.Tensor) -> torch.Tensor:
    return 0.5 * x * (1 + torch.erf(x / math.sqrt(2)))
```

```python
# Alternative: Fast approximation
def gelu_approx(x: torch.Tensor) -> torch.Tensor:
    return 0.5 * x * (1 + torch.tanh(math.sqrt(2 / math.pi) * (x + 0.044715 * x ** 3)))
```

```python
# Verify
x = torch.tensor([-2., -1., 0., 1., 2.])
print("Input: ", x)
print("Exact: ", gelu(x))
print("Approx:", gelu_approx(x))
```
