# 🔴 Solution: SwiGLU Activation

```python
import torch
import torch.nn as nn
```

```python
# ✅ SOLUTION

class SwiGLU(nn.Module):
    def __init__(self, d_in: int, d_out: int):
        super().__init__()
        self.gate_proj = nn.Linear(d_in, d_out, bias=False)
        self.up_proj = nn.Linear(d_in, d_out, bias=False)
    
    def forward(self, x: torch.Tensor) -> torch.Tensor:
        gate = self.gate_proj(x)
        up = self.up_proj(x)
        # SiLU = x * sigmoid(x)
        swish = gate * torch.sigmoid(gate)
        return swish * up
```

```python
# Verify
glu = SwiGLU(d_in=32, d_out=32)
x = torch.randn(2, 6, 32)
out = glu(x)
print(f"Output shape: {out.shape}")
```
