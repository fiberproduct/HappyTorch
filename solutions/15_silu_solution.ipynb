# 🟢 Solution: SiLU (Swish) Activation

```python
import torch
```

```python
# ✅ SOLUTION

def silu(x: torch.Tensor) -> torch.Tensor:
    return x * torch.sigmoid(x)
```

```python
# Verify
x = torch.tensor([-2., -1., 0., 1., 2.])
print("Input: ", x)
print("Output:", silu(x))
```
